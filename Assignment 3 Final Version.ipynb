```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classificati## DATA PREPARATIONon_report
from sklearn.metrics import confusion_matrix
```

## DATA PREPARATION

```python
train = pd.read_csv('train.csv', index_col = 'PassengerId')
test = pd.read_csv('test.csv', index_col = 'PassengerId')## DATA PREPARATION
```

```python
train.info()
test.info()
```

```python
train['Age'].fillna(train['Age'].mean(), inplace = True)
test['Age'].fillna(test['Age'].mean(), inplace = True)
train['Embarked'].fillna("S", inplace = True)
train['Survived'] = train['Survived'].astype(str)
```

```python
train.info()
test.info()
```

```python
group_names = ['Young', 'Middle-Aged', 'Senior']
train['Age Categories'] =pd.cut(train['Age'], 3, labels = group_names)
test['Age Categories'] = pd.cut(test['Age'], 3, labels = group_names)
```

```python
train.head()
```

```python
test.head()
```

```python
train = train.drop(train.columns[[2, 4, 7, 9]], axis = 1)
test = test.drop(test.columns[[1, 3, 6, 8]], axis = 1)
```

```python
train.head()
```

```python
test.head()
```

```python
# FOR SEX CATEGORY
sexdummytrain = pd.get_dummies(train['Sex'])
train = pd.concat([train, sexdummytrain], axis = 1, sort = True)

sexdummytest = pd.get_dummies(test['Sex'])
test = pd.concat([test, sexdummytest], axis = 1, sort = True)

# FOR EMBARKED CATEGORY
embarkedtrain = pd.get_dummies(train['Embarked'])
train = pd.concat([train, embarkedtrain], axis = 1, sort = True)

embarkedtest = pd.get_dummies(test['Embarked'])
test = pd.concat([test, embarkedtest], axis = 1, sort = True)

# FOR AGE CATEGORY
agetrain = pd.get_dummies(train['Age Categories'])
train = pd.concat([train, agetrain], axis = 1, sort = True)

agetest = pd.get_dummies(test['Age Categories'])
test = pd.concat([test, agetest], axis = 1, sort = True)
```

```python
train.head()
```

```python
test.head()
```

```python
train = train.drop(train.columns[[2, 6, 7]], axis = 1)
test = test.drop(test.columns[[1, 5, 6]], axis = 1)
```

```python
train.head()
```

```python
test.head()
```

Since we have applied the same data manipulation techniques to both train and test data, the below will be reserved only for creating the prediction model based on train data.

## CREATING THE MODEL FROM TRAIN DATA

```python
train.head()
```

```python
features = ['Pclass', 'SibSp', 'Parch', 'Fare','female', 'male', 'C', 'Q', 'S', 'Young', 'Middle-Aged', 'Senior']
target = ['Survived']
```

```python
X = train[features]
Y = train[target]
X_train, X_testing, Y_train, Y_testing = train_test_split(X, Y, test_size = 0.4, random_state = 42)
```

```python
model = DecisionTreeClassifier(min_samples_leaf = 5)
model.fit(X_train, Y_train)
```

## USING PREDICTION MODEL ON TRAIN'S SPLIT DATA

```python
Prediction = model.predict(X_testing)
Prediction
```

```python
# evaluate result  

print("Accuracy:", accuracy_score(Y_testing, Prediction, normalize = True, sample_weight = None))
print("Classification Report:", classification_report(Y_testing, Prediction))
print("Confusion Matrix:", confusion_matrix(Y_testing, Prediction))
```
